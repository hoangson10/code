```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
time = ctrl.Antecedent(np.arange(10,121,1),'time')
rice = ctrl.Antecedent(np.arange(10,1001,1),'rice')
power = ctrl.Consequent(np.arange(10,101,1),'power')
time['1'] = fuzz.trimf(time.universe,[10,10,37.5])
time['2'] = fuzz.trimf(time.universe,[10,37.5,65])
time['3'] = fuzz.trimf(time.universe,[37.5,65,92.5])
time['4'] = fuzz.trimf(time.universe,[65,92.5,120])
time['5'] = fuzz.trimf(time.universe,[92.5,120,120])
rice['1'] = fuzz.trimf(rice.universe,[100,100,325])
rice['2'] = fuzz.trimf(rice.universe,[100,325,550])
rice['3'] = fuzz.trimf(rice.universe,[325,550,775])
rice['4'] = fuzz.trimf(rice.universe,[550,775,1000])
rice['5'] = fuzz.trimf(rice.universe,[775,1000,1000])
power['1'] = fuzz.trimf(power.universe,[10,28,46])
power['2'] = fuzz.trimf(power.universe,[28,46,64])
power['3'] = fuzz.trimf(power.universe,[46,64,82])
power['4'] = fuzz.trimf(power.universe,[64,82,100])
rule1= ctrl.Rule(time['1']&rice['1'],power['1'])
rule2= ctrl.Rule(time['1']&rice['2'],power['1'])
rule3= ctrl.Rule(time['1']&rice['3'],power['2'])
rule4= ctrl.Rule(time['1']&rice['4'],power['3'])
rule5= ctrl.Rule(time['1']&rice['5'],power['3'])
rule7= ctrl.Rule(time['2']&rice['1'],power['1'])
rule8= ctrl.Rule(time['3']&rice['1'],power['2'])
rule9= ctrl.Rule(time['4']&rice['1'],power['3'])
rule10= ctrl.Rule(time['5']&rice['1'],power['3'])
rule11= ctrl.Rule(time['2']&rice['2'],power['1'])
rule12= ctrl.Rule(time['2']&rice['3'],power['2'])
rule13= ctrl.Rule(time['2']&rice['4'],power['3'])
rule14= ctrl.Rule(time['2']&rice['5'],power['4'])
rule15= ctrl.Rule(time['3']&rice['2'],power['2'])
rule16= ctrl.Rule(time['3']&rice['3'],power['3'])
rule17= ctrl.Rule(time['3']&rice['4'],power['4'])
rule18= ctrl.Rule(time['3']&rice['5'],power['4'])
rule19= ctrl.Rule(time['4']&rice['2'],power['3'])
rule20= ctrl.Rule(time['4']&rice['3'],power['3'])
rule21= ctrl.Rule(time['4']&rice['4'],power['4'])
rule22= ctrl.Rule(time['4']&rice['5'],power['4'])
rule23= ctrl.Rule(time['5']&rice['2'],power['3'])
rule24= ctrl.Rule(time['5']&rice['3'],power['4'])
rule25= ctrl.Rule(time['5']&rice['4'],power['4'])
rule26= ctrl.Rule(time['5']&rice['5'],power['4'])



powering_rule=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26])
powering= ctrl.ControlSystemSimulation(powering_rule)
powering.input['time']=56
powering.input['rice']=600
powering.compute()
print(powering.output['power'])
power.view(sim=powering)
```
